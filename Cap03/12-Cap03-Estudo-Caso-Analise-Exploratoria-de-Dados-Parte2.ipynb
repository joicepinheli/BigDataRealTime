```python
# Importando o arquivo csv e alterando o tipo dos dados definidos na coluna Data de Objetc para data
df = pd.read_csv('data/dataframe_saved_v1.csv', parse_dates = ['Data'])
```

```python
# Retornando as 5 primeiras linhas do DataFrame
df.head()
```

```python
# Retornando os tipos dos dados contidos no DataFrame
df.dtypes
```

```python
# Alterando o nome das colunas
cols = ['Data', 'Distancia', 'Tempo']
df.columns = cols
df.head()
```

```python
# Definindo a coluna Data como índice
df.set_index('Data', inplace = True)
```

```python
# Retornando as 5 primeiras linhas do DataFrame
df.head()
```

```python
# Criando um Plot
df.plot()
```

```python
# Função para converter a coluna de duração no tempo em segundos
def calcula_total_segundos(time):
    if time is np.nan:
        return np.nan
    hrs, mins, seconds = str(time).split(':')
    seconds = int(seconds) + 60 * int(mins) + 60 * 60 * int(hrs)
    return seconds
```

```python
# Aplicando nova função nos dados da coluna Tempo do DataFrame e criando uma nova coluna "Segundos"
df['Segundos'] = df.Tempo.map(calcula_total_segundos)
```

```python
# Retornando as 10 primeiras linhas do DataFrame
df.head(10)
```

```python
# Obtendo o resumo estatístico dos dados
df.describe()
```

```python
# Preenchendo os valores NaN com zeros
df.fillna(0).describe()
```

```python
# Criando uma nova colunas Minutos
df['Minutos'] = df['Segundos'].map(lambda x: x/60)
```

```python
# Preenchendo os valores NaN com zeros
df.fillna(0).describe()
```

```python
# Criando um Gráfico Scatterplot
df.plot(x = 'Distancia', y = 'Minutos', kind = 'scatter')
```

```python
# Chamando o método CORR para analizar a correlação entre as variáveis
df.corr()
```

```python
# Chamando o método CORR para analizar a correlação entre as variáveis
df.corr(method = 'spearman')
```

```python
# Chamando o método CORR para analizar a correlação entre as variáveis
df.corr(method = 'kendall')
```

```python
# Criando um Boxplot para analisar outliers
df.boxplot('Minutos', return_type = 'axes')
```

```python
# Criando uma nova variável para analisar os minutos por Kilometro
df['Min_Por_Km'] = df['Minutos'] / df['Distancia']
```

```python
# Preenchendo os valores NaN com zeros
df.fillna(0).describe()
```

```python
# Criando um histograma, para visualizar a distribuição de frequência da nova variável
df.hist('Min_Por_Km')
```

```python
# Alterando a quantidade de colunas do histograma
df.hist('Min_Por_Km', bins = 20)
```

```python
# Salvando o gráfico
df.hist('Min_Por_Km', bins = 20, figsize = (10,8))
plt.xlim((5, 11))
plt.ylim((0, 12))
plt.title("Histograma Minutos por KM")
plt.grid(False)
plt.savefig('imagens/hist_minutos_por_km.png')
```
